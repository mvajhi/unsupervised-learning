```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
distances = []
dimensions = range(2, 51)

for d in dimensions:
    x = np.random.randint(0, 1000, size=d)
    y = np.random.randint(0, 1000, size=d)

    distance = np.linalg.norm(x - y)
    distances.append(distance)
```

```python
plt.plot(dimensions, distances, marker='o')
```

فاصله با افزایش ابعاد زیاد می شود. این موضوع باعث می شود که داده ها پراکنده تر شوند و سخت تر بتوانیم داده ها را نزدیک و همسایه تعریف کنیم.
