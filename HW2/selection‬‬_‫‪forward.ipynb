## Remove outlire

ستون آخر اشتباه و ناقص است آن را حذف می کنیم و همچنین ستون id بی اهمیت است آن را هم حذف می کنیم. همچنین چون در صورت سوال گفته شده داده های پرت را حذف کنیم آن ها را حذف می کنیم در حالت عادی معقول نیست این کارو در حالت کلی انجام بدیم باید علت دادگان پرت را بدانیم و با آگاهی آن ها را حذف کنیم.
می شود داده های پرت را با گزینش خاصی جدا کرد ولی در حالت کلی ۱.۵ طول جعبه در نمودار جعبه ای خارج از جعبه به بعد را داده پرت در نظر می گیرد. ماهم آنها را حذف می کنیم.

```python
df = df.drop(columns=['id', 'Unnamed: 32'])
```

```python
features = [col for col in df.columns if col != 'diagnosis' and np.issubdtype(df[col].dtype, np.number)]
df_cleaned = df.copy()


inlier_mask = pd.Series(True, index=df.index)
for col in features:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    inlier_mask &= (df[col] >= lower_bound) & (df[col] <= upper_bound)

df_cleaned = df[inlier_mask]
print(np.sum(~inlier_mask))
```

## Classification

```python
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
df_cleaned['diagnosis'] = LabelEncoder().fit_transform(df_cleaned['diagnosis'])

X = df_cleaned.drop(columns=['diagnosis'])
y = df_cleaned['diagnosis']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
model = xgb.XGBClassifier()
model.fit(X_train, y_train)

y_pred_train = model.predict(X_train)
y_pred = model.predict(X_test)

accuracy_train = accuracy_score(y_train, y_pred_train)
accuracy_test = accuracy_score(y_test, y_pred)
print("test acc: ", accuracy_train)
print("test acc: ", accuracy_test)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

## Feature Selection

```python
from sklearn.feature_selection import SequentialFeatureSelector
```

```python
MAX_FEATURES = 10

for k in range(1, MAX_FEATURES + 1):
    print("".join(['-']*40))
    print(k)

    sfs = SequentialFeatureSelector(estimator=model,
                                  n_features_to_select=k,
                                  direction='forward')
    sfs.fit(X_train, y_train)


    sf_mask = sfs.get_support()
    sf_names = X_train.columns[sf_mask]
    print(sf_names)

    X_train_sfs = sfs.transform(X_train)
    X_test_sfs = sfs.transform(X_test)

    model_sfs = xgb.XGBClassifier()
    model_sfs.fit(X_train_sfs, y_train)

    y_pred_sfs = model_sfs.predict(X_test_sfs)
    y_pred_train_sfs = model_sfs.predict(X_train_sfs)

    accuracy_train = accuracy_score(y_train, y_pred_train_sfs)
    accuracy_test = accuracy_score(y_test, y_pred_sfs)
    print("train acc: ", accuracy_train)
    print("test acc: ", accuracy_test)
    print(confusion_matrix(y_test, y_pred_sfs))
```
