```python
from ultralytics import YOLO

wandb.init(project="vehicle_detection", name="yolov8s_run")

model = YOLO('yolov8s.pt')

results = model.train(data='/content/My-First-Project-1/data.yaml', epochs=50, imgsz=640, project='vehicle_detection', name='yolov8s_run')

wandb.finish()
```

```python
wandb.init(project="vehicle_detection", name="yolov8m_run")

model_m = YOLO('yolov8m.pt')

results_m = model_m.train(data='/content/My-First-Project-1/data.yaml', epochs=50, imgsz=640, project='vehicle_detection', name='yolov8m_run')
wandb.finish()
```

```python
wandb.init(project="vehicle_detection", name="yolov8l_run")

model_l = YOLO('yolov8l.pt')

results_l = model_l.train(data='/content/My-First-Project-1/data.yaml', epochs=50, imgsz=640, project='vehicle_detection', name='yolov8l_run')
wandb.finish()
```

```python
import pandas as pd

metrics_s = results.results_dict
metrics_m = results_m.results_dict
metrics_l = results_l.results_dict

comparison_data = {
    'Model': ['Small', 'Medium', 'Large'],
    'mAP50-95': [
        metrics_s['metrics/mAP50-95(B)'],
        metrics_m['metrics/mAP50-95(B)'],
        metrics_l['metrics/mAP50-95(B)']
    ],
    'mAP50': [
        metrics_s['metrics/mAP50(B)'],
        metrics_m['metrics/mAP50(B)'],
        metrics_l['metrics/mAP50(B)']
    ],
    'Precision': [
        metrics_s['metrics/precision(B)'],
        metrics_m['metrics/precision(B)'],
        metrics_l['metrics/precision(B)']
    ],
    'Recall': [
        metrics_s['metrics/recall(B)'],
        metrics_m['metrics/recall(B)'],
        metrics_l['metrics/recall(B)']
    ]
}

df = pd.DataFrame(comparison_data)

df = df.round(3)
display(df)
```

```python
from IPython.display import Image, display
import glob

print("Small")
display(Image(filename='/content/vehicle_detection/yolov8s_run/results.png', width=800))

print("Medium")
display(Image(filename='/content/vehicle_detection/yolov8m_run/results.png', width=800))

print("Large")
display(Image(filename='/content/vehicle_detection/yolov8l_run/results.png', width=800))
```

```python
import matplotlib.pyplot as plt
import numpy as np

models = df['Model']
map_scores = df['mAP50-95']

plt.figure(figsize=(10, 6))
bars = plt.bar(models, map_scores)
plt.ylabel('mAP50-95')

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval, f'{yval:.3f}', va='bottom', ha='center')

plt.show()
```

```python
import cv2
from google.colab.patches import cv2_imshow

model_s = YOLO('/content/vehicle_detection/yolov8s_run/weights/best.pt')
model_m = YOLO('/content/vehicle_detection/yolov8m_run/weights/best.pt')
model_l = YOLO('/content/vehicle_detection/yolov8l_run/weights/best.pt')

test_image_paths = glob.glob('/content/My-First-Project-1/test/images/*.jpg')[:3]

for image_path in test_image_paths:
    results_s = model_s(image_path);
    results_m = model_m(image_path);
    results_l = model_l(image_path);

    result_image_s = results_s[0].plot()
    result_image_m = results_m[0].plot()
    result_image_l = results_l[0].plot()

    result_image_s_rgb = cv2.cvtColor(result_image_s, cv2.COLOR_BGR2RGB)
    result_image_m_rgb = cv2.cvtColor(result_image_m, cv2.COLOR_BGR2RGB)
    result_image_l_rgb = cv2.cvtColor(result_image_l, cv2.COLOR_BGR2RGB)
    original_image_rgb = cv2.cvtColor(cv2.imread(image_path), cv2.COLOR_BGR2RGB)


    fig, axs = plt.subplots(1, 4)

    axs[0].imshow(original_image_rgb)
    axs[0].set_title('Original')
    axs[0].axis('off')

    axs[1].imshow(result_image_s_rgb)
    axs[1].set_title('Small')
    axs[1].axis('off')

    axs[2].imshow(result_image_m_rgb)
    axs[2].set_title('Medium')
    axs[2].axis('off')

    axs[3].imshow(result_image_l_rgb)
    axs[3].set_title('Large')
    axs[3].axis('off')

    plt.tight_layout()
    plt.show()
```
